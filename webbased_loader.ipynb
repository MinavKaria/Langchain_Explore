```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/associate-vendeur-h-f-cdi-30heures-semaine/job/R-67594")
docs = loader.load().pop().page_content

print(docs)
# docs[0]
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template=PromptTemplate.from_template("Extract the company names from : {document}")

prompt_template.invoke(
    {"document": docs}
)
```
