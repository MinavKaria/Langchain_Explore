```python
from langchain_groq import ChatGroq
import os
import getpass

if not os.environ.get("GROQ_KEY"):
  os.environ["GROQ_KEY"] = getpass.getpass("Enter API key for Groq: ")

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0,
    api_key=os.environ["GROQ_KEY"]
    # other params...
)

res=llm.invoke("What is Apple ?")
print(res)
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/senior-manager-product-management-distribution-and-outbound-transportation-itc/job/R-66230")
docs = loader.load().pop().page_content

print(docs)
# docs[0]
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template=PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBPAGE
    {document}
    ### INSTRUCTIONS
    The scraped text is from career's page from a website.
    Your task is to extract the job posting and return them in JSON format containing the following keys: 'role', 'skills' and 'description'.
    Only return the JSON object without any additional text or preamble.
    """
)

# chain = LLMChain(
#     llm=llm,
#     prompt=prompt_template
# )

chain_extract=prompt_template | llm
res=chain_extract.invoke({"document": docs})
print(res.content)
```
